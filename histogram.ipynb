```python
import matplotlib.pyplot as plt
import pickle
```

```python
with open('./pickle/Brownspot_lap0.pickle', 'rb') as handle:
    Brownspot_lap0 = pickle.load(handle)
    
with open('./pickle/Healthy_lap0.pickle', 'rb') as handle:
    Healthy_lap0 = pickle.load(handle)
```

```python
plt.subplot(1, 2, 1)
plt.hist(Brownspot_lap0, bins=50,normed=True)
plt.subplot(1, 2, 2)
plt.hist(Healthy_lap0, bins=50,normed=True)
plt.savefig("Histogram/red_lap.png")
plt.show()
```

```python
plt.hist(Brownspot_lap0, bins=50,normed=True)
#plt.savefig("Histogram/red_lap_BrownSpot.png")
plt.show()
plt.hist(Healthy_lap0, bins=50,normed=True)
#plt.savefig("Histogram/red_lap_Healthy.png")
plt.show()
```
